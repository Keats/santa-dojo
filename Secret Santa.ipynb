```python
import random

def iter_pairings(participants):
    ps = participants
    random.shuffle(ps)
    ps += participants[:1]
    return zip(ps, ps[1:])
```

```python
attendees = open('attendee-checkin-list.txt', encoding='utf8').read().splitlines()[16::11]
attendees = [a for a in attendees if a]
assert len(set(attendees)) == len(attendees)
attendees
```

```python
from collections import OrderedDict
pairings = list(iter_pairings(attendees))
print(pairings)
d = OrderedDict(pairings)
```

```python
def s_eq(a, b):
    a = set(a)
    b = set(b)
    assert a == b, '%r != %r (mismatch is %r)' % (a, b, a ^ b)

s_eq(d.keys(), d.values())
```
